## Duplicated Primary Keys

Each table should have a unique **Primary Key**. Let's do a sanity check to make sure they are indeed unique.

```python
#Printing Sales ID

print(sales['%KEY'].value_counts(dropna=False).head(3),
      item_master['Item Number'].value_counts(dropna=False).head(3),
      sales_rep['Sales Rep ID'].value_counts(dropna=False).head(3),
      cities['City Code'].value_counts(dropna=False).head(3),
      customers['Customer Number'].value_counts(dropna=False).head(3),
      sep='\n\n'
     )
```

Oops! It turns out that customer number `10021911` is actually duplicated in the `customers` source data table. This could be due to a misconfigured database that allowed a duplicate primary key to sneak into the database... it happens!

```python
customers[ customers['Customer Number'] == 10021911 ]
```

The difference appears to be in the customer's name being capitalized or not, and a different city code. Unfortunately in the absence of further information it is difficult here to determine why the duplicate Customer Number has happened. 

A duplicate Customer Number will throw a wrench in our joins later when we stitch tables up. So, the easiest thing to do here is to note the discrepancy, and delete the extra customer, or assign its customer number to a value that doesn't exist.

In this case, the next customer number in the sequence, `10021912`, doesn't exist. So we'll set `Ping` to that number. 

```python
customers[ customers['Customer Number'] == 10021912 ].shape
```

```python
customers.iloc[468]['Customer Number']
```

```python
customers.iloc[468]['Customer Number'] = 10021912
```

```python
customers.at[468, 'Customer Number'] = 10021912
```

The way to set a particular value is to use the `at` syntax. There are other ways as well, but for simplicity it is recommended to use `at` for setting a single value. ([docs](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.at.html))

```python
customers[ (customers['Customer Number'] == 10021911) | (customers['Customer Number'] == 10021912) ]
```

```python
customers['Customer Number'].value_counts(dropna=False).head(3)
```

## Join

Join all your tables together and store the joined result into a dataframe named `cs`. You'll need this for the Reporting and Visualization section below. 

Use the data dictionary for guidance.

Make sure the shape of the joined table makes sense!

```python
#Join

print(sales.shape)

cs = sales.merge(item_master, how='left', on='Item Number') \
          .merge(sales_rep,   how='left', left_on='Sales Rep Number', right_on='Sales Rep ID') \
          .merge(customers,   how='left', on='Customer Number') \
          .merge(cities,      how='left', on='City Code')

print(cs.shape)
```

```python
pd.set_option('display.max_columns', 34)
cs.head()
```

It turns out we don't need a bunch of columns, so let's drop 'em so things run faster. How did I know what column is unwanted? I asked my instructor :)

```python
cs.drop(['%KEY', 'Customer Number', 'Invoice Date', 'Invoice Number', \
         'Item Desc', 'Item Number', 'Order Number', 'Promised Delivery Date', 'Sales Rep Number', \
         'Manager', 'Manager Number', 'Path', 'Product Line', \
         'Sales Rep Name1', 'Sales Rep Name2', 'Sales Rep Name3', 
         'Sales Rep Number', 'Sales Rep ID', 'Customer', \
         'City Code', 'City', 'Region', 'Latitude', 'Longitude', 'Desc'], axis=1, inplace=True)
```

```python
cs.head(10)
```

## 1. Gross Sales by Product Type (Top 10)

```python
#GroupBy
df_gs = cs.groupby('Product Type').sum() \
                                 .sort_values(by='GrossSales', ascending=False) \
                                 .head(10) \
                                 .apply(lambda x: x/1_000_000) \
                                 .reset_index()

#Plots
plt.figure(figsize=(15,7))
ax = sns.barplot(x='Product Type', y='GrossSales', data=df_gs, palette = 'viridis_r');
ax.set(ylabel='Gross Sales [MM $USD]', title ='Gross Sales [MM $USD] by Product Type');
```

We can see that Vegetables, Meat and Canned Shrimp contributed much more than the rest combined. These products are moving at a quick rate.

## 2. Average Margin (%) by Product Group

```python
#GroupBy
df_am = cs[['Product Group', 'MarginPct']].groupby('Product Group').mean() \
                                          .sort_values(by='MarginPct', ascending = False) \
                                          .reset_index()

#Plot
plt.figure(figsize=(15,7))
ax = sns.barplot(x='MarginPct', y='Product Group', data=df_am, palette = 'viridis_r');
ax.set(xlabel='Average Margin [%]', title='Average Margin [%] by Product Group');
```

Minimum average margin is contributed by Alcoholic becerages at a rate of 40% Average margin. Which is not bad when compared to the highest margin which is about 62%.

## 3. Top 10 Sales Reps and their Total Sales and Sales Quantity

(Formatted as a table)

```python
sales_by_rep = cs[['Sales Rep Name', 'Sales Qty', 'Sales']] \
              .groupby('Sales Rep Name').sum() \
              .sort_values(by='Sales', ascending=False) \
              .head(10)

sales_by_rep['Sales'] = sales_by_rep['Sales'].apply(lambda x: x/1_000_000)
sales_by_rep.rename(columns={ 'Sales': 'Sales [MM $USD]' }, inplace=True)
sales_by_rep
```

Judy has comparatively lesser number of sales quantity while comparing the top three, however, she has managed to provide the highest sales. Looks like she took quality over quantity for real.

## 4. Scatter Plot of Average Margin [%] vs Average Gross Sales

```python
#GroupBy
am_vs_ags = cs[ ['Product Sub Group', 'MarginPct', 'GrossSales'] ] \
           .groupby('Product Sub Group').mean() \
           .sort_values(by='GrossSales', ascending=False) \
           .head(20)

am_vs_ags.reset_index(inplace=True)
```

```python
#Scatterplot
plt.figure(figsize=(13,6))

ax = sns.scatterplot(x='MarginPct', y='GrossSales', data=am_vs_ags, hue='Product Sub Group', palette = 'cubehelix')
ax.set(xlabel='Average Margin [%]', ylabel='Average Gross Sales [$USD]', title='Average Margin vs Average Gross Sales');
```

```python
#Gross Sales over 8000
am_vs_ags[ am_vs_ags['GrossSales'] > 8000 ]
```

## 5. Total Sales by Month

```python
#Total Sales By Month
sales_by_date = cs[['Date', 'Sales']].copy()
sales_by_date.set_index(sales_by_date['Date'], inplace=True)
sales_by_date.drop(['Date'], axis=1, inplace=True)
sales_by_date
```

```python
#Let's truncate it to month:
total_sales_by_month = sales_by_date.groupby(sales_by_date.index.to_period('M')) \
                                    .sum() \
                                    .apply(lambda x: x/1_000_000) \
                                    .reset_index()
total_sales_by_month['Sale']  = np.round(total_sales_by_month.Sales, 2)
total_sales_by_month.head()
```

```python
#Let's do a plot from matplotlibrary, the OG:

total_sales_by_month[['Date','Sale']].plot(kind='bar', width = 0.40, color='lightblue', figsize=(15,5))
plt.ylabel('Sales [MM $USD]')
plt.xlabel('Month')
plt.title('Total Sales by Month');
total_sales_by_month['Sales'].plot(secondary_y=True, color = 'r');
```

## 6. Total Sales by Month, Year over Year

```python
#Total Sales by Month, partition by year. 
total_sales_by_month = cs[['Date', 'Sales']].groupby([cs['Date'].dt.year, cs['Date'].dt.month]) \
                                            .sum() \
                                            .apply(lambda x: x/1_000_000)
total_sales_by_month
```

```python
total_sales_by_month.unstack()
```

```python
tsbmut = total_sales_by_month.unstack().transpose()
tsbmut
```

```python
tsbmut.set_index(tsbmut.index.droplevel(level=0), inplace=True)
tsbmut
```

```python
tsbmut.plot(figsize=(15,5), xticks=tsbmut.index)
plt.ylabel('Sales [MM $USD]')
plt.xlabel('Month')
plt.title('Monthly Sales, YOY');
```

We could see that that the sales have been consistent. Except for the Sales at 2012, which has gone super low. Could be becayse of the storm? Or maybe because the country was recovering from the recession from 2007 - 2009.
